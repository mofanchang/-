```python
# 載入基礎套件與代稱
import numpy as np
import matplotlib.pyplot as plt
```

```python
def mean_absolute_error(y, yp):
    """
    計算 MAE
    Args:
        - y: 實際值
        - yp: 預測值
    Return:
        - mae: MAE
    """
    mae = MAE = sum(abs(y - yp)) / len(y)
    return mae

# 定義 mean_squared_error 這個函數, 計算並傳回 MSE
def mean_squared_error(y,yp):
    diff = abs(y - yp)
    # Take the square of the difference
    squared_diff = diff ** 2
    # Calculate the mean of the squared difference
    mse = sum(squared_diff) / len(y)
    return mse
```

MAE : Mean Absolute Error / 平均絕對誤差
常用在財務方面, 在數學又稱 L1 損失, MAE 就是將每次測量的絕對誤差取絕對值後求的平均值, 用來觀察預測值誤差的測量是否要調整。

模型擬合的值與觀察值之間差的平方除以減去模型中參數數目的歷程點數的總和。 模型中的參數數目會從歷程點的數目中減去，使其與無偏差的模型變異估計值一致。

MSE : Mean Square Error / 均方誤差
optimal constant : target mean
均方誤差 (MSE) 是最常用的回歸損失函數，計算方法是求預測值與真實值之間距離的平方和. 由於有平方, 懲罰偏離真實比較重, 因此適合梯度計算. 在數學又稱 L2 損失, MSE 可以評價資料的變化程度，MSE 的值越小，說明預測模型描述實驗資料具有更好的精確度。

```python
from math import sqrt

def mean_squared_error(y, yp):
  # Calculate the difference between the true and predicted values
  diff = y - yp
  
  # Take the absolute value of the difference
  abs_diff = abs(diff)
  
  # Calculate the mean of the squared difference
  mse = sum(abs_diff) / len(y)
  
  # Take the square root of the mean squared error
  rmse = sqrt(mse)
  
  return rmse
```

```python
# 與範例相同, 不另外解說
w = 3
b = 0.5
#  Generate an array of 101 evenly spaced values from 0 to 100
x_lin = np.linspace(0, 100, 101)
# Calculate the y values using a linear function with random noise added
y = (x_lin + np.random.randn(101) * 5) * w + b

plt.plot(x_lin, y, 'm.', label = 'data points')
#'b.' argument specifies that the points should be plotted as blue dots.
plt.title("Assume we have data points")
plt.legend(loc = 2)
plt.show()
```

```python
https://www.cnblogs.com/IvyWong/p/9916791.html
# 與範例相同, 不另外解說
y_hat = x_lin * w + b
plt.plot(x_lin, y, 'b.', label = 'data')
plt.plot(x_lin, y_hat, 'r--', label = 'prediction')
plt.title("Assume we have data points (And the prediction)")
plt.legend(loc = 2)
plt.show()
```

```python
# 執行 Function, 確認有沒有正常執行
MSE = mean_squared_error(y, y_hat)
MAE = mean_absolute_error(y, y_hat)
print("The Mean squared error is %.3f" % (MSE))
print("The Mean absolute error is %.3f" % (MAE))
```
